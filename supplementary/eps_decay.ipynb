```python
import matplotlib.pyplot as plt
```

```python
def plot_eps_decay(eps_start, eps_end, eps_decay, n_iter):
    eps = eps_start
    track = [eps]
    for _ in range(n_iter):
        eps = eps * eps_decay if eps > eps_end else eps_end
        track.append(eps)
    plt.figure(figsize=(5, 5))
    plt.plot(track)
    plt.show()

```

```python
max_i = 100000
gamma = 0.98
alpha = 0.05
eps_start = 0.99
eps_end = 0.0
eps_decay = 0.99995

plot_eps_decay(eps_start, eps_end, eps_decay, max_i)
```
