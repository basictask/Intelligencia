# Segéd dokumentum az $\varepsilon$ hiperparaméter hangolásához
## Használat:
    1. Ki kell tölteni a paramétereket 
    2. Le kell futtatni a függvényt, ami ábrázolja a paraméter csökkenését a tanítási iterációkon keresztül
    3. Úgy kell hangolni a paramétereket, hogy az ügynöknek legyen lehetősége a tanulási folyamat elején felfedezni, a vége felé pedig kizsákmányolni a megszerzett tudását

```python
import matplotlib.pyplot as plt
```

```python
def plot_eps_decay(eps_start, eps_end, eps_decay, n_iter):
    eps = eps_start
    track = [eps]
    for _ in range(n_iter):
        eps = eps * eps_decay if eps > eps_end else eps_end
        track.append(eps)
    plt.figure(figsize=(5, 5))
    plt.plot(track)
    plt.show()
```

```python
max_i = 20000
eps_end = 0.0  # Felfedezési ráta kezdőértéke
eps_start = 1.0  # Felfedezési ráta végértéke
eps_decay = 0.9998  # Felfedezési ráta iterációnként ennyivel fog csökkenni
eps = eps_start

plot_eps_decay(eps_start, eps_end, eps_decay, max_i)
```
